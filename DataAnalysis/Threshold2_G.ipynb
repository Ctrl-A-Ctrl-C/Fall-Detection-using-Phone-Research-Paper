```python
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# assign directory
directory = "C:/Users/phuon/EDX/EPQDataAnalysis/_Extracted Data 2/"
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        df = pd.read_csv(f, delimiter = ",")
        data_AX = ((df["AX"])**2 + (df["AY"])**2 + (df["AZ"])**2)**1/2
        data_GX = ((df["GX"])**2 + (df["GY"])**2 + (df["GZ"])**2)**1/2
        data_AX_max = data_AX.max()
        data_GX_max = data_GX.max()

        if (data_AX_max >= 1.6) and (data_GX_max >= 8):
            print(f + " ----- User has fallen")
        else:
            print(f + " ----- User has not fallen")
```

```python
cf_matrix = [[18, 8], [4, 0]]
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Confusion Matrix of Magnitude \n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Positive','Negative'])
ax.yaxis.set_ticklabels(['True','False'])

## Display the visualization of the Confusion Matrix.
plt.show()
```
