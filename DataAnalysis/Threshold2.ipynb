```python
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# assign directory
directory = "C:/Users/phuon/EDX/EPQDataAnalysis/_Extracted Data/"
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        df = pd.read_csv(f, delimiter = ",")
        data = ((df["AX"])**2 + (df["AY"])**2 + (df["AZ"])**2)**(1/2)
        data_max = data.max()
        print(data_max)

        if (data_max >= 1.8):
            print(f + " ----- User has fallen")
        else:
            print(f + " ----- User has not fallen")
```

```python
cf_matrix = [[18, 0], [4, 8]]
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Confusion Matrix of Magnitude \n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Fall','Not Fall'])
ax.yaxis.set_ticklabels(['Fall','Not Fall'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
TV_5_2 = pd.read_csv("C:/Users/phuon/EDX/EPQDataAnalysis/_Extracted Data/filesTV_5_2.csv", delimiter = ",")

TV_5_2 = TV_5_2.interpolate()

TV_5_2["Vector"] = ((TV_5_2["AX"])**2 + (TV_5_2["AY"])**2 + (TV_5_2["AZ"])**2)**(1/2)

TV_5_2["Vector"].plot(kind='line', #to produce a stacked area plot, each column must either be all pos. or all neg.
             stacked=False, #area plots are automatically stacked. Pass this for it to be unstacked. NaN would be defaulted as 0
             figsize=(20, 10), # pass a tuple (x, y) size
             fontsize = 14
             )

plt.title('TV_Dropping_2', fontsize = 20)
plt.ylabel('Accelerometer Data', fontsize = 20)
plt.xlabel('Samples', fontsize = 20)
plt.legend(fontsize = 20)

TV_5_2.head()
```
