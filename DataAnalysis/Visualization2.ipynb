```python
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
BD_1_1 = pd.read_csv("C:/Users/phuon/EDX/EPQDataAnalysis/_Extracted Data/filesBD_1_1.csv", delimiter = ",")

data = ((BD_1_1["AX"])**2 + (BD_1_1["AY"])**2 + (BD_1_1["AZ"])**2)**1/2

df_BD_1_1 = pd.DataFrame(data, columns=['Magnitude'])
```

```python
df_BD_1_1.head()
```

```python
df_BD_1_1.plot(kind='line', #to produce a stacked area plot, each column must either be all pos. or all neg.
             stacked=False, #area plots are automatically stacked. Pass this for it to be unstacked. NaN would be defaulted as 0
             figsize=(20, 10), # pass a tuple (x, y) size
           )

plt.title('Magnitude of BD_1_1')
plt.ylabel('Accelerometer Data')
plt.xlabel('Time/s')

plt.show()
```
