```python
%matplotlib inline
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

BD_1_1 = pd.read_csv("C:/Users/phuon/EDX/EPQDataAnalysis/filesBD_1_1.csv", delimiter = ",")
BD_1_1.tail()
```

```python
headerList = ["AX", "AY", "AZ", "GX", "GY", "GZ"]
BD_1_1.to_csv("C:/Users/phuon/EDX/EPQDataAnalysis/filesBD_1_1.csv", header = headerList, index=False)
BD_1_1 = pd.read_csv("C:/Users/phuon/EDX/EPQDataAnalysis/filesBD_1_1.csv", delimiter = ",")
BD_1_1.head()
```

```python
BD_1_1.drop(["GX", "GY", "GZ"], axis=1, inplace=True)
BD_1_1.head()
```

```python
BD_1_1["AX"].astype(float)
```

```python
BD_1_1["AX"] = pd.to_numeric(BD_1_1["AX"], errors='coerce')
BD_1_1["AY"] = pd.to_numeric(BD_1_1["AY"], errors='coerce')
BD_1_1["AZ"] = pd.to_numeric(BD_1_1["AZ"], errors='coerce')

BD_1_1.plot(kind='line', #to produce a stacked area plot, each column must either be all pos. or all neg.
             stacked=False, #area plots are automatically stacked. Pass this for it to be unstacked. NaN would be defaulted as 0
             figsize=(20, 10), # pass a tuple (x, y) size
           )

plt.title('BD_Walking_1')
plt.ylabel('Accelerometer Data')
plt.xlabel('Time/s')

plt.show()
```
